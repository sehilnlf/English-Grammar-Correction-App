### Spelling check

```python
import requests
import difflib

raw_url = 'https://raw.githubusercontent.com/dwyl/english-words/master/words_alpha.txt'
response = requests.get(raw_url)

if response.status_code == 200:
    content = response.text
else:
    print('Failed to retrieve the file from GitHub:', response.status_code)

# Stopwords
words = content.split('\n')
words = list(map(lambda x: x[:-1], words))

def spelling_check(text):
    text = text.lower()
 
    for word in text.split():
        if word not in words:
            
            # Use difflib to find similar words within the list of correct words
            # n = number of closest word
            # cutoff = tỷ lệ giống nhau
            suggestions = difflib.get_close_matches(word, words, n=5, cutoff=0.8)
            if suggestions:
                print(f"Word '{word}' might be misspelled. Suggestions: {suggestions}")
            else:
                print(f"No suggestions found for the word '{word}'. It might be misspelled.")

# Example usage
spelling_check("helo and welcom to the new worl")
```
